```python
import pandas as pd
```

```python
df = pd.read_csv("games.csv", index_col=0)
```

```python
df["date"] = pd.to_datetime(df['date'])
df.dtypes
```

```python
#covert home/ away to cat 
df['venue_code'] = df['venue'].astype('category').cat.codes #convert string to category then category into numbers
df['opponent_code'] = df['opponent'].astype('category').cat.codes #convert string to category then category into numbers

df.head()
```

```python
df['hour'] = df['time'].str.replace(':.+', "", regex=True).astype('int')
```

```python
df['day_code'] = df['date'].dt.day_of_week
```

```python
#convert the target to a boolean 
df['target']= (df['result'] =='W').astype('int')
df
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = df[df['date'] < '2022-01-01']
test = df[df['date'] > '2022-01-01']
```

```python
predictors = ['venue_code', 'opponent_code', 'hour', 'day_code']
```

```python
rf.fit(train[predictors], train['target'])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test['target'], preds)
acc
```

```python
combined = pd.DataFrame(dict(actual = test['target'], predictions = preds))
```

```python
pd.crosstab(index=combined['actual'], columns=combined['predictions'])
```

```python
from sklearn.metrics import precision_score
precision_score(test['target'], preds)
```
